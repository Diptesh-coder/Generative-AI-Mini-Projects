```python
from groq import Groq
import base64 # It is package that provide module to convert jpg to base64 image
import os # importing into operating system
import pandas as pd
import json
from google.colab import userdata, files
```

GROQ API Key initialization

```python
# NOTE: Insert your Groq API key as an environment variable or use a config file. Do not hardcode secrets in code.
client = Groq(api_key="API key")  # API key removed for security
```

```python
# Defining the function to encode the image
def encode_image(image_path):
  with open(image_path, "rb") as image_file:
    return base64.b64encode(image_file.read()).decode('utf-8')
```

```python
folder_path = "/content/Dataset"
all_results = []
```

```python
print("Starting OCR")
for filename in os.listdir(folder_path):
    if filename.endswith((".jpg",".jpeg",".png")):
      print(f"Processing {filename}...")
      full_path = os.path.join(folder_path, filename)
      base64_image = encode_image(full_path)
      completion = client.chat.completions.create( model="meta-llama/llama-4-scout-17b-16e-instruct", messages=[ { "role": "user", "content": [ {"type": "text", "text": "Extract these fields as JSON: 'Merchant', 'Date', 'Total', 'Category'. If you can't find one, use 'N/A'."}, {"type": "image_url", "image_url": {"url": f"data:image/jpeg;base64,{base64_image}"}} ] } ], response_format={"type": "json_object"} )

      result_data = json.loads(completion.choices[0].message.content)
      result_data['Source_File'] = filename
      all_results.append(result_data) # Appending all the results

print("OCR Completed")
```

```python
df = pd.DataFrame(all_results)
```
