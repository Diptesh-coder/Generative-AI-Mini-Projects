```python
from groq import Groq
import base64 # It is package that provide module to convert jpg to base64 image
import os # importing into operating system

# NOTE: Insert your Groq API key as an environment variable or use a config file. Do not hardcode secrets in code.
client = Groq(api_key="")  # API key removed for security
```

```python
# Defining the function to encode the image
def encode_image(image_path):
  with open(image_path, "rb") as image_file:
    return base64.b64encode(image_file.read()).decode('utf-8')
```

```python
# Setting the path to our image
image_path = "/content/Screenshot 2025-11-15 205222.png"
```

```python
# Getting the base64 string
base64_image = encode_image(image_path)

client = Groq(api_key="")  # API key removed for security
chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": [
                {"type": "text", "text": "You are an expert ocr model. Give me the exact text that is written in the image. Also give me word count of the image."},
                {
                    "type": "image_url",
                    "image_url": {
                        "url": f"data:image/jpeg;base64,{base64_image}",
                    },
                },
            ],
        }
    ],
    model="meta-llama/llama-4-scout-17b-16e-instruct",
)
```

Printing the result

```python
print(chat_completion.choices[0].message.content)
```
