2. Importing Groq, base64 and os

```python
from groq import Groq
import base64
import os
```

3. Function to encode the image

```python
def encode_image(image_path):
  with open(image_path, "rb") as image_file:
    return base64.b64encode(image_file.read()).decode('utf-8')
```

4. Path to your image

```python
image_path = "/content/Receipt.jpg"
```

5. Getting the base64 string

```python
base64_image = encode_image(image_path)

client = Groq(api_key="YOUR_GROQ_API_KEY")

chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": [
                {"type": "text", "text": "Extract the contents in tabular form just like an accountant."},
                {
                    "type": "image_url",
                    "image_url": {
                        "url": f"data:image/jpeg;base64,{base64_image}",
                    },
                },
            ],
        }
    ],
    model="meta-llama/llama-4-scout-17b-16e-instruct",
)
```

6. Printing the output

```python
print(chat_completion.choices[0].message.content)
```
