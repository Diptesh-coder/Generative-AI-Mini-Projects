Importing Groq, os and json

```python
from groq import Groq
import base64
import os
import json
```

Initializing the Groq client

```python
client = Groq(api_key="YOUR_GROQ_API_KEY")
```

Specifying the path to the audio file

```python
filename = "/content/Harvard/harvard.wav"
```

Opening the audio file

```python
with open(filename, "rb") as file:
    file_content = file.read()
```

Creating a transcription of the audio file and printing

```python
with open(filename, "rb") as audio_file:
    transcription = client.audio.transcriptions.create(
          file=audio_file, # Required audio file
          model="whisper-large-v3-turbo",
          prompt="Specify context or spelling",
          response_format="verbose_json",
          timestamp_granularities = ["word", "segment"],
          language="en",
          temperature=0.0
        )
print(json.dumps(transcription, indent=2, default=str)) #Printing the transcription
```

Generating 5 line summary

```python
summary_lines = []
for i, segment in enumerate(transcription.segments):
    if i < 5:
        summary_lines.append(segment['text'].strip())
    else:
        break

for line in summary_lines:
    print(line)
```
